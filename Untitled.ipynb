```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
from sklearn.cluster import KMeans, DBSCAN
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage.data import astronaut
from npGraph import Graph, visuals
```

```python
img = img_as_float(cv2.imread("./Images/image.png", 0))/255
plt.imshow(img)
```

```python
# img = img_as_float(astronaut())
segments = slic(img, n_segments=1000, compactness=0.001, sigma=1, start_label=1)
new_img = np.zeros_like(img)
for i in np.unique(segments):
    x = np.where(segments==i)
    new_color = np.sum([img[i, j] for i,j in zip(x[0], x[1])], axis=0)/len([img[i, j] for i,j in zip(x[0], x[1])])
    new_img[x] = new_color
fig, axs = plt.subplots(3, figsize=(10,30))
axs[0].imshow(img)
axs[1].imshow(new_img)
axs[2].imshow(mark_boundaries(img*255, segments))
```

```python
edge_list = []
for i in range(segments.shape[0]):
    for j in range(segments.shape[1]):
        x = set(np.unique(segments[i:i+2, j:j+2]))
        if len(x)==2:
            if x not in edge_list:
                edge_list.append(x)
edge_list = [list(i) for i in edge_list]
edge_list
```
