```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
from sklearn.cluster import KMeans, DBSCAN
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage.data import astronaut
from npGraph import Graph, visuals
import seaborn as sns
from sklearn.cluster import SpectralClustering
```

```python
img = img_as_float(cv2.imread("./Images/0940.control.40.a.tif", 0))
plt.imshow(img)
```

```python
sample = []
win_size = 1000
counter = 0
for i in range(0, img.shape[0], 10):
    for j in range(0, img.shape[1], 10):
        x = img[i:i+win_size, j:j+win_size]
        counter += 1
        if counter==1000:
            sample.append(x)
print(counter)
plt.imshow(sample[0])
```

```python
img = sample[0]
segments = slic(img, n_segments=10000, compactness=.1, sigma=1, start_label=0)
new_img = np.zeros_like(img)
mean_pixel_values = dict()
segment_id = {}
for i in np.unique(segments):
    x = np.where(segments==i)
    new_color = np.sum([img[i, j] for i,j in zip(x[0], x[1])], axis=0)/len([img[i, j] for i,j in zip(x[0], x[1])])
    new_img[x] = new_color
    mean_pixel_values[i] = float(new_color)
    segment_id[i] = i
fig, axs = plt.subplots(3, figsize=(90,30))
axs[0].imshow(img)
axs[1].imshow(new_img)
axs[2].imshow(mark_boundaries(img, segments))
```

```python
num_sp = int(np.unique(segments).shape[0])
print(num_sp)
```

```python
edge_list = []
for i in range(segments.shape[0]):
    for j in range(segments.shape[1]):
        x = set(np.unique(segments[i:i+2, j:j+2]))
        if len(x)==2:
            if x not in edge_list:
                edge_list.append(x)

def edge_weight(p1, p2, sigma=0.1):
    return math.exp(-((p2-p1)**2)/(2*(sigma**2)))

edge_list = [list(i)+[edge_weight(mean_pixel_values[list(i)[0]], mean_pixel_values[list(i)[1]], sigma=0.001)] for i in edge_list]
sns.kdeplot([i[-1] for i in edge_list])
```

```python
G = Graph.from_edges(edge_list)
```

```python
for i in G.nodes[0]:
    G.adj[i][np.where(G.adj[i]<0.6)] = 0
```

```python
for i in G.nodes[0]:
    x = np.where(G.adj[i]>0)[0]
    print(x)
    for node in x:
        segment_id[node] = i
```

```python
np.unique(list(segment_id.values()))
```

```python
visuals.vis(G, "graph")
```

```python
num_clusters = 20
clustering = SpectralClustering(num_clusters, affinity='precomputed', n_init=100).fit(G.adj)
labels = clustering.labels_
```

```python
new_img = np.zeros_like(segments).astype(float)
for i in range(num_clusters):
    x = np.mean([mean_pixel_values[j] for j in np.where(labels==i)[0]])
    new_img[np.where(labels==i)] = x
```
